```python
num_clusters = 5
from sklearn.cluster import KMeans
# Create a KMeans clustering model
kmeans = KMeans(n_clusters=num_clusters)

# Fit the model to your data
kmeans.fit(bdf)

# Get the cluster labels for each data point in your dataset
labels = kmeans.labels_

# Create a new DataFrame with the cluster labels
clustered_df = pd.concat([bdf, pd.Series(labels, name='cluster')], axis=1)

# Sample from each cluster
sample_size = 4
clustered_sample = pd.concat([clustered_df[clustered_df['cluster'] == i].sample(sample_size,replace=True) for i in range(num_clusters)])
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

n_samples=384
strata_size = 100
stratifier = "Class"
n_splits = 1  # we only need one split since we're taking a sample of the entire dataframe

# Initialize the StratifiedShuffleSplit object with the desired parameters
sss = StratifiedShuffleSplit(n_splits=n_splits, test_size=strata_size,train_size=n_samples, random_state=42)

# Get the indices of the rows in the dataframe that are selected for the sample
for train_index, test_index in sss.split(bdf, bdf[stratifier]):
    sample_indices = test_index
    break

# Use the sample_indices to extract the rows from the original dataframe
sample_df = bdf.iloc[sample_indices, :]
```
